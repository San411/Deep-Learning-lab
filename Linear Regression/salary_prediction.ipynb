# **Implementation of linear regression**


**Predicting Salary based on experience**

```
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Importing the file
path = '/content/drive/MyDrive/6th Sem/Deep Learning/Assignment 3/Salary_Data.csv'
dataFrame = pd.read_csv(path)
dataFrame
```

```
# Checking for null values
dataFrame.isnull().sum()
```

```
# describing the data
dataFrame.describe()
```

```
# Assigning the dependent and Independent variables
X = dataFrame.iloc[:, :-1].values
y = dataFrame.iloc[:, -1:].values
X, y
```

```
# Splitting the dataset into train and test values (20%)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```
# Importing Linear Regression model from Scikit learn
from sklearn.linear_model import LinearRegression
simpleLR = LinearRegression()
# Training the model
simpleLR.fit(X_train, y_train)
```

```
# Predicting the salary
y_pred = simpleLR.predict(X_test)
y_pred
```

```
# Predicting the salary for 10 years of experience
simpleLR.predict([[10]])
```

```
# Actual and predicted values
print(np.concatenate((y_test.reshape(len(y_test), 1), y_pred.reshape(len(y_pred), 1)), 1))
```

```
# Evaluating the model
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(y_pred, y_test), mean_squared_error(y_pred, y_test), np.sqrt(mean_squared_error(y_pred, y_test)), simpleLR.score(X_test, y_test)
```
